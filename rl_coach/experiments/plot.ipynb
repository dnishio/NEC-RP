```python
df = pd.read_csv('./pong-rp/25_10_2018-18_19/simple_rl_graph.main_level.main_level.agent_0.csv')
df2 = pd.read_csv('./pong-sparse/25_10_2018-19_36/simple_rl_graph.main_level.main_level.agent_0.csv')
df = df[['Total steps','Evaluation Reward']]
df2 = df2[['Total steps','Evaluation Reward']]
df = pd.rolling_mean(df.dropna(), 1)
df2 = pd.rolling_mean(df2.dropna(), 1)
df = df.dropna()
df2 = df2.dropna()
```

```python
ax = df.plot(x=df.columns[0])
ax.set_ylabel("Rewards")
ax2 = df2.plot(ax=ax, x=df2.columns[0], title='Pong', grid=True)
ax2.legend(['Gaussian RP', 'Sparse RP'])
```
